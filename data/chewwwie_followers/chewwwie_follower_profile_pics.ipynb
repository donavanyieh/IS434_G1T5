```python
import pandas as pd
import csv
```

```python
chewwwie_filtered_followers_df = pd.read_csv(r'chewwwie_filtered_followers.csv')
chewwwie_filtered_followers_df
```

```python
chewwwie_filtered_followers_df_img = chewwwie_filtered_followers_df['imageUrl']
chewwwie_filtered_followers_df_img
```

```python
chewwwie_filtered_followers_img_list = chewwwie_filtered_followers_df_img.to_list()
chewwwie_filtered_followers_img_list 
```

```python
import requests
import shutil

count=1
for url in chewwwie_filtered_followers_img_list:
    url_filename = url.split("/")[-1].split('?')[0]  + ".png"
    print(url_filename)
    print(count)


    r = requests.get(url, stream=True)

    if r.status_code == 200:
        r.raw.decode_content = True

        with open(url_filename, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
            
            print("Image successfully Downloaded:", url_filename)
    else:
        print("Unable to download the image")
    
    count+=1
```
