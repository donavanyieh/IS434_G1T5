# JSON to CSV 

```python
#merge json files
import glob, os, json
import pandas as pd

json_dir = r'C:\Users\hengd\Desktop\Year_2\Y2S2\IS434 _SAA\Project\IGScrap\fukudonsg'

json_pattern = os.path.join(json_dir, '*.json')
file_list = glob.glob(json_pattern)

dfs = []
for file in file_list:
    with open(file) as f:
        json_data = pd.json_normalize(json.loads(f.read()))
        json_data['site'] = file.rsplit("/", 1)[-1]
    dfs.append(json_data)
df = pd.concat(dfs)
```

```python
df.head()
```

```python
df.dropna(axis=1)
```

```python
df.to_csv("json.csv")
```
