This notebook serves to <b>extract and download first 5 photos of all of Chewwie's filtered followers</b>.

Images are saved to:<br><b>/IS434_G1T5/data/followers_posts/chewwie_followers_images</b>

Filtered followers refer to <b>Chewwie's</b> followers whose profile meets the following criteria:
<ol>
    <li>Public account</li>
    <li>Number of posts > 5 </li>
    <li>Number of followers > 100 </li>
    <li>Followers > (Following * 2/3)</li>
</ol>

Data was retrieved using phantombuster.

## Import Dependencies

```python
import pandas as pd
import csv
import requests
import shutil
```

## Retrieve images

```python
chewwwie_followers_posts_df = pd.read_csv("../../data/followers_posts/chewwwie_followers_posts.csv")
chewwwie_followers_posts_df.head()
```

```python
images_list = chewwwie_followers_posts_df['imgUrl'].to_list()
images_list[:3]
```

```python
count=1
for url in images_list:
    url_filename = url.split("/")[-1].split('?')[0]  + ".png"
    print(url_filename)
    print(count)


    r = requests.get(url, stream=True)

    if r.status_code == 200:
        r.raw.decode_content = True

        with open(url_filename, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
            
            print("Image successfully Downloaded:", url_filename)
    else:
        print("Unable to download the image")
    
    count+=1
```
