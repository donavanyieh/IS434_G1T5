```python
import pandas as pd
import csv
```

```python
chewwwie_followers_posts_df1 = pd.read_csv(r'chewwwie_followers_posts1.csv')
chewwwie_followers_posts_df1
```

```python
chewwwie_followers_posts_df1_img = chewwwie_followers_posts_df1['imgUrl']
chewwwie_followers_posts_df1_img
```

```python
chewwwie_followers_posts_df1_img_list = chewwwie_followers_posts_df1_img.to_list()
chewwwie_followers_posts_df1_img_list = chewwwie_followers_posts_df1_img_list[1:]
chewwwie_followers_posts_df1_img_list
```

```python
import requests
import shutil

count=1
for url in chewwwie_followers_posts_df1_img_list:
    url_filename = url.split("/")[-1].split('?')[0]  + ".png"
    print(url_filename)
    print(count)


    r = requests.get(url, stream=True)

    if r.status_code == 200:
        r.raw.decode_content = True

        with open(url_filename, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
            
            print("Image successfully Downloaded:", url_filename)
    else:
        print("Unable to download the image")
    
    count+=1
```
