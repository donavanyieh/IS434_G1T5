```python
fukudon_filtered_followers_df = pd.read_csv(r'result_imageurl.csv')
fukudon_filtered_followers_df
```

```python
fukudon_filtered_followers_df_img = fukudon_filtered_followers_df['imgUrl']
fukudon_filtered_followers_df_img
```

```python
fukudon_filtered_followers_img_list = fukudon_filtered_followers_df_img.to_list()
fukudon_filtered_followers_img_list 
```

```python
import requests
import shutil

count=1
for url in fukudon_filtered_followers_img_list:
    url = str(url)
    url_filename = url.split("/")[-1].split('?')[0]  + ".png"
    print(url_filename)
    print(count)

    try :
        r = requests.get(url, stream=True)

        if r.status_code == 200:
            r.raw.decode_content = True

            with open(url_filename, 'wb') as f:
                shutil.copyfileobj(r.raw, f)

                print("Image successfully Downloaded:", url_filename)
        else:
            print("Unable to download the image")
    except :
        pass
    
    count+=1
```
